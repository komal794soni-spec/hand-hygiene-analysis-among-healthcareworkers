INTERPRETATION AND SUMMARY OF HAND HYGIENE COMPLIANCE 

This Notebook evaluates the Hand Hygiene compliance among health care workers in different aspects such as:
1. Overall Compliance (5 Moments and 7 Steps)
2. Performance of HCW's in Different Categories (Doctor, Nurse, GDA/HK, Other)
3. Preference of HCW'S for Cleaning hands (Handrub / Handwash)
4. Compliance in Critical and Semi-critical care areas
5. Areawise Compliance among HCW'S
6. Moment wise Compliance
7. visual and Statistical comparison

The main aim of this Study is to analyse the Compliance and identify the Highest and lowest performing Groups and vulnerable areas to understand the patterns and control the infection risk in a tertiary care hospital

7.3 STATISTICAL ANALYSIS

I. HCW AND 5 MOMENTS COMPLIANCE (Chi Square)

```python
'''
Chi-square value:  33.60246465606425
Degree of freedom:  3
p-value:  2.4034472645234014e-07
'''
```

INTERPRETATION

1. non uniform compliance among HCW's
2. The difference is not likely to be due to a chance, it indicates real behavioural variation among health care worker categories

II. HAND RUB VS HAND WASH PREFERENCES AMONG HCW'S (Chi square)

```python
'''
Chi-square value:  23.804044344536592
Degree of freedom:  3
p-value:  2.7447072779652652e-05
'''
```

Interpretation

1. There is variation in pattern of preferences among HCW's
2. It does not show any random fluctuation

III. CRITICAL AND SEMICRITICAL COMPLIANCE (Z-Test)
