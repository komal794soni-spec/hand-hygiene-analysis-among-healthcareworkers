This Notebook consists of visual analysis of the data obtained from the publication in the form of basic Bar and line charts.

5.1 IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

5.2 IMPORTING CSV FILE

```python
# USING read.csv() Functions {to read data from CSV file and load the data in to pandas DataFrame}

compliance = pd.read_csv("overall_compliance_HH.csv")
moment_step_compliance = pd.read_csv("Moments_Steps_compliance_HH.csv")
preference = pd.read_csv("Preference_HR_HW.csv")
care_area_compliance = pd.read_csv("Critical_Semicritical_Area_Compliance.csv")
areawise_compliance = pd.read_csv("Area_Wise_Compliance.csv")
momentwise = pd.read_csv("Moment_Wise_Compliance.csv")
```

5.3 GRAPH SHOWS THE COMPLIANCE OF HEALTHCARE WORKERS DURING 5 MOMENTS WITH7 STEPS OF HAND HYGIENE

```python

plt.figure()
plt.plot(moment_step_compliance["HCW"], moment_step_compliance["Moments(%)"],marker = "o", label = "5 Moments %")
plt.plot(moment_step_compliance["HCW"], moment_step_compliance["Steps(%)"],marker = "o",  label = "7 Steps %")
plt.xlabel("Healthcare Workers")
plt.ylabel("Compliance (%) of 5 Moments with 7 Steps")
plt.title("Compliance by HCW Category during 5 Moments of Hand Hygiene")
plt.legend()
plt.show()
```

5.4 THE GRAPH SHOWS THE PREFERENCE OF HAND RUB V/S HAND WASH AMONG HCW'S 

```python
plt.figure()
plt.plot(preference ["HCW"], preference ["HR (%)"],marker = "o",  label = "Hand Rub")
plt.plot(preference ["HCW"], preference ["HW (%)"],marker = "o",  label = "Hand Wash")
plt.xlabel("Healthcare Workers")
plt.ylabel("Preference (%)")
plt.title("PREFERENCE OF HAND WASH AND HAND RUB AMONG HCWS")
plt.legend()
plt.show()
```

5.5 THE GRAPH SHOWS THE COMPLIANCE OF HCW'S IN CRITICAL AND SEMI-CRITICAL CARE AREAS

```python
plt.figure()
plt.plot(care_area_compliance["HCW"], care_area_compliance["C (%)"],marker = "o",  label = "Critical")
plt.plot(care_area_compliance["HCW"], care_area_compliance["SC (%)"],marker = "o",  label = "Semi-Critical")
plt.xlabel("Healthcare Workers")
plt.ylabel("Compliance (%)")
plt.title("COMPLIANCE OF HCW'S IN CRITICAL AND SEMI-CRITICAL CARE AREAS")
plt.legend()
plt.show()
```

5.6 THE GRAPH SHOWS THE AREA WISE COMPLIANCE OF HCW'S IN CRITICAL AND SEMICRITICAL AREAS

```python
plt.figure()
plt.plot(areawise_compliance["Areas"], areawise_compliance["Doctors (%)"],marker = "o", label = "Doctor")
plt.plot(areawise_compliance["Areas"], areawise_compliance["Nurses (%)"], marker = "o", label = "Nurse")
plt.plot(areawise_compliance["Areas"], areawise_compliance["GDA/HK (%)"], marker = "o", label = "GDA/HK")
plt.plot(areawise_compliance["Areas"], areawise_compliance["Other (%)"], marker = "o", label = "Other")
plt.xlabel("Healthcare Workers")
plt.ylabel("Compliance (%)")
plt.title("AREAWISE COMPLIANCE OF HCW'S IN CRITICAL AND SEMI-CRITICAL CARE AREAS")
plt.legend()
plt.grid()
plt.show()
```

5.7 THE GRAPH SHOWS THE MOMENT WISE COMPLAINCE 

```python
momentwise["Moments"] = ["M1", "M2", "M3", "M4", "M5"]
momentwise

plt.figure()
plt.bar(momentwise["Moments"], momentwise["Compliance (%)"], label = "Compliance")
plt.xlabel("Moments")
plt.ylabel("Compliance (%)")
plt.title("# MOMENT WISE COMPLIANCE")
plt.legend()
plt.show()
```
